# User and Group Management

- Manage Local User Account
  - When creating a new user, files in `/etc/skel` will be copied to the new user home directory
  - `passwd` can be used to modify user passwords
  - `userdel` can be used to remove a user account. The home directory is not deleted by default. You
    need to use `userdel --remove <username>`
  - /etc/passwd stores user information (login shell, home directory, hashed password, uid, gid, username, etc.)
  - to create a system account, use `useradd --system <username>`
  - chage can be used to expire a password or configure password expiration policies for a user account
  - use `gpasswd` to add users to groups
    ```
    [root@centos01 sysctl.d]# gpasswd -a john developers
    Adding user john to group developers  
    ```
  - users can have one primary group and several secondary groups
  - you cannot remove a group if it is someones primary group
  - Password aging:
    - can be changed using the `chage` command
    - Password aging information is stored in `/etc/shadow` file
    - Password aging default policies are stored in `/etc/login.defs`
    - To list current password aging settings for a user:
      - `chage -l <username>`
  
- Manage System Wide Environment Profiles
  - add env vars to ~/.bashrc
  - system wide env vars are stored in /etc/environment

 - Configure User Resource Limits
   - edit `/etc/security/limits.conf` to configure user resource limits
     - this file is well documented!
   - type `ulimit -a` to view your current limits
      ```
      [limited@lfcs3 ~]$ ulimit -a
      core file size          (blocks, -c) 0
      data seg size           (kbytes, -d) unlimited
      scheduling priority             (-e) 0
      file size               (blocks, -f) unlimited
      pending signals                 (-i) 127694
      max locked memory       (kbytes, -l) 64
      max memory size         (kbytes, -m) unlimited
      open files                      (-n) 1024
      pipe size            (512 bytes, -p) 8
      POSIX message queues     (bytes, -q) 819200
      real-time priority              (-r) 0
      stack size              (kbytes, -s) 8192
      cpu time               (seconds, -t) unlimited
      max user processes              (-u) 3
      virtual memory          (kbytes, -v) unlimited
      file locks                      (-x) unlimited
      ```

  - Manage User Privileges
    - Users of the `wheel` group have superuser privs on the system
    - `/etc/sudoers` specifies who can use sudo on the system
      - use `visudo` to modify this file
      - ```
        ## Same thing without a password
        # %wheel        ALL=(ALL)       NOPASSWD: ALL 
        ^^^^^^^^        ^^^  ^^^
        <User or Group> <Hostname or IP> <User> <Commands that can be executed with sudo>
        ```
